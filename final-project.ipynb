```python
from scripts.NN2 import NN
from scripts.io import encode_data
from scripts.k_fold import k_fold_learning_rate, get_k_split_data
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
encoded_pos, encoded_neg = encode_data('data/rap1-lieb-positives.txt','data/yeast-upstream-1k-negative.fa')
sets = get_k_split_data(10, encoded_pos, encoded_neg)
learning_rates, training_error, testing_error = k_fold_learning_rate(10,sets)
```

```python
plt.figure()
plt.plot(learning_rates,training_error)
plt.xlabel('learning rate')
plt.ylabel('error')
plt.title('Training error 10-fold cross validation')
plt.show()

plt.figure()
plt.plot(learning_rates,testing_error)
plt.xlabel('learning rate')
plt.ylabel('error')
plt.title('Test error for 10-fold cross validation')
plt.show()
```
